# Core Data Structures

```python
# Series: A Series is a one-dimensional array-like object. It can hold any data type like integers, strings, floats, etc. 

data = [10, 20, 30, 40, 50]
series = pd.Series(data)

print(series)
```

```python
# You can also assign custom indexes:

series = pd.Series(data, index=['a', 'b', 'c', 'd', 'e'])
print(series)
```

```python
# Data Frame: A DataFrame is a two-dimensional data structure that can hold different types of data in columns (like a table).

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Score': [85, 90, 95]
})

print(df)
```

```python
df.loc[0]            # label-based

# df refers to a Pandas DataFrame.
# .loc[] is a label-based indexing method. This means it accesses rows and columns by their labels (index names and column names), rather than their integer positions.
# [0]indicates that the row with the label 0 is being selected.
```

```python
df.iloc[0]           # integer-location based

# df: Represents a Pandas DataFrame object.
# .iloc: Is a purely integer-location based indexing method used for selection by position. It accepts various inputs for indexing, including single integers, lists of integers, slice objects, and boolean arrays.
# [0]: Specifies the integer position of the row to be selected. Since Python uses zero-based indexing, 0 refers to the first row of the DataFrame.
```
